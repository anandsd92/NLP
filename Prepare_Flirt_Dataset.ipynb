```python

def PreProcessing(filename):
    df = pd.read_csv(filename,header=None,on_bad_lines='skip',encoding='utf8')
    if(len(df.columns)==2):
            df=df.drop(0)
            df=df.dropna(axis=0)
            df.columns=['Date','Text']
    else:
            df=df.drop([2,3],axis=1)
            df=df.dropna(axis=0)
            df=df.drop(0)
    df.columns=['Date','Text']
    df["Date"]=pd.to_datetime(df["Date"],errors='coerce',dayfirst=True)
    df = df.dropna(subset=["Date"])
    df = df[~df["Text"].str.contains("<Media omitted>", na=False)]
    #df.drop(['Text1','Text2'],axis=1,inplace=True)
    #print(df.head(20))
    print(df.shape[0])
    return df
```

```python
def SplitTextIntoColumns(df):
    Message = df["Text"].str.split("-",n=1,expand=True)
    Message.columns=["Time","Chat"]
    df["Time"]=Message["Time"]
    print(Message.head())
    #Message1=Message["Chat"].str.split(":",n=1,expand=True)
    Message1 = Message["Chat"].str.extract(r'^(.*?):\s*(.*)$')
    print(Message1.head())
    # Assign columns
    df["Name"] = Message1[0].fillna("System")   # default if no name
    df["Text"] = Message1[1].fillna(Message["Chat"])  # fallback to full chat if no message part
    df=df[["Date","Time","Name","Text"]]
    return df
```

```python
def Classify_Flirt(text):
   flirt_values= [

    # Compliments
    "your smile",
    "looked really great",
    "you looked great",
    "nicest person",

    # Wanting to be close
    "sitting next to you",
    "look forward to sitting next to you",
    "save me a seat",
    "hoped you'd save me a seat",

    # Enjoying their company
    "talking with you",
    "favorite part of the day",
    "make every school day more interesting",

    # Emotional indicators
    "get happy whenever i see you",
    "happy whenever i see you",
    "reason i don't mind coming to school",

    # Special attention
    "partners for the project",
    "look forward to",
    "secretly hoped"

   ]
   if text is None:
      return "Non-Flirt"

   text = str(text).lower().strip()
   return "Flirt" if any(kw.lower().strip() in text for kw in flirt_values) else "Non-Flirt"
```
