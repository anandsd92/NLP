```python
# Import text vectorizers from scikit-learn:
# - CountVectorizer: converts text into word counts
# - TfidfVectorizer: converts text into TF-IDF scores (importance of words)
# - HashingVectorizer: converts text into fixed-length feature vectors using hashing
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, HashingVectorizer
from sklearn.decomposition import NMF
```

```python
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')# Assign the value for max and min allowance stopwords
dtm=tfidf.fit_transform(df['Text'])# Transforming commentText to tfdif vector values
```

```python
print(type(dtm))
```

```python
nmf_model = NMF(n_components=5,random_state=42)#Assigning number of topics
nmf_model.fit(dtm)#Tranforming Transformed commentText formula to NMF .This stores the output of 5 topic modelling topics
#cc=nmf_model.components_
```

```python
results=[]
for index, topic in enumerate(nmf_model.components_):
    print(type(topic))
    result=([tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    results.append([index,result])
```

```python
print(type(results))
print(results)
```

```python
topic_results = nmf_model.transform(dtm)
df['Topic'] = topic_results.argmax(axis=1)#Assign respective topics to each commentTent
```
